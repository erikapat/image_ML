## **Opening Image Files in a Notebook**

```python
### Install cv2 = opencv
!pip install opencv-python
```

```python
import cv2
img = cv2.imread('data/00-puppy.jpg')
img
```

The image will be loaded by openCV as a numpy array, with the color sorted as BGR. For this reasons appear in blue color

```python
img_bgr = cv2.imread('data/00-puppy.jpg')
plt.imshow(img_bgr)
```

Matplotlib's plot expects an RGB image so, for a correct display of the image, it is necessary to swap those channels. 
This operation can be done either by using openCV conversion functions cv2.cvtColor() or by working directly with the numpy array.

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

Now, we can se what happen with IMREAD_GRAYSCALE

```python
img_gray = cv2.imread('data/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

## **GRAY COLORS**

```python
img_gray = cv2.imread('data/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
```

## **Resize Images**

```python
## Resize Images
img_rgb.shape
# width, height, color channels
```

```python
img =cv2.resize(img_rgb,(1300,275))
plt.imshow(img)
```

### **By ratio**

```python
w_ratio = 0.5
h_ratio = 0.5
new_img =cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
plt.imshow(new_img)
```

### **Flipping Images**

```python
%matplotlib inline
# Along central x axis
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
# Along central y axis
new_img = cv2.flip(new_img,1)
plt.imshow(new_img)
```

```python
# Along both axis
new_img = cv2.flip(new_img,-1)
plt.imshow(new_img)
```

## **Saving Image Files**

```python
type(new_img)
cv2.imwrite('data/my_new_picture.jpg',new_img)
```

## **Larger Displays in the Notebook**

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.imshow(new_img)
```
